```tablegen
%reset
def ins;
def Val;     // 只是占个位用的叶子符号，名字可以叫别的，比如 ArgType、Foo 都行

class Base<string name, dag input = (ins)> {
    string op = name;
    dag arg = !con(input, (ins Val:$c));
}

def case : Base<"case", (ins Val:$a, Val:$b)>;
```

```tablegen
%reset
// 定义一个模板，用来描述“带名字的占位类型”
class ArgWithName<string kind> {
  string k = kind;
}

// 一个 dag 头用的标记
def node;

// 一个类，接受一个 dag 做参数
class Holder<dag d> {
  dag stuff = d;
}

// 现在直接实例化 ArgWithName<"..."> 放进 dag
def UseIt : Holder<(node ArgWithName<"foo">:$a,
                         ArgWithName<"bar">:$b)>;

```
